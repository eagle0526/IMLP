```python
#練習 - 彙總球隊薪資 繪製成質方圖
data = pd.read_csv(r"D:\python機器學習\Unit02\data\NBA_season1718_salary.csv")
# data["Tm2"] = data["Tm"]
# data

#用樞紐分析表，記得一定要用reset_index()，要不然會無法使用該undex值

salary_by_team = data.groupby(by="Tm").sum().reset_index()
salary_by_team = salary_by_team.sort_values(by="season17_18", ascending=False)
salary_by_team

tm_top10 = salary_by_team[1:10]

tm = tm_top10["Tm"]
salary_sum = tm_top10["season17_18"]
sns.barplot(tm, salary_sum, data=tm_top10)
sns.color_palette("hls", 10)
sns.set(style="darkgrid")
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(plt.hist,"tip")
```

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend()
```

```python
#這個很重要，就是我一直在找的 - 可以用散點圖做 多種顏色的設定

pal = dict(Lunch="seagreen", Dinner="gray")
g = sns.FacetGrid(tips, hue="time", palette=pal, height=8)
g.map(plt.scatter, "total_bill", "tip", s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.add_legend()
```
